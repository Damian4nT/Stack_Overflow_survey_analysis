### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

dbase=df.copy()

print('null values before cleaning',dbase['DatabaseWantToWorkWith'].isnull().sum())
dbase=dbase.dropna(subset=['DatabaseWantToWorkWith'])
print('\n\nnull values after cleaning',dbase['DatabaseWantToWorkWith'].isnull().sum())

dbase['dbase']=dbase['DatabaseWantToWorkWith'].str.split(';')
dbase=dbase.explode('dbase')                                                                                # Split responses containing multiple databases into separate rows so each database can be analyzed individually

top5db=dbase['dbase'].value_counts().head(5)

top5db.plot(
    kind='pie'
    ,figsize=(10,8)
    ,autopct='%1.1f%%'
    ,startangle=90
    ,labels=top5db.index
    ,pctdistance=0.7
    ,title=' Top 5 Databases Respondents Want to Work With'
)
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here

devtype=df.copy()

print('null values before cleaning',devtype['DevType'].isnull().sum())
devtype=devtype.dropna(subset=['DevType'])
print('\n\nnull values after cleaning',devtype['DevType'].isnull().sum())

top5devs=devtype['DevType'].value_counts().head(5)

top5devs.plot(
    kind='pie'
    ,figsize=(12,8)
    ,autopct='%1.1f%%'
    ,startangle=90
    ,labels=top5devs.index
    ,pctdistance=0.8
    ,title='top 5 developer roles'
)
plt.show()
```
