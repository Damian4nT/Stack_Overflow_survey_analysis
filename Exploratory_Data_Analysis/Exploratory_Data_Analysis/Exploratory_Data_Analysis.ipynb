#### Step 3: Handling Missing Data


Identify and manage missing values in critical columns such as `Employment`, `JobSat`, and `RemoteWork`. Implement a strategy to fill or drop these values, depending on the significance of the missing data.


```python
## Write your code here
jobsat_df=df.copy()
jobsat_df[['Employment','JobSat','RemoteWork']].isnull().sum()        # checking columns that are important for exercise
```

```python
def before_clean(df_col):
    print('null values before cleaning', df_col.isnull().sum())

def after_clean(df_col):
    print('null values after cleaning', df_col.isnull().sum())
```

```python
jobsat_df=jobsat_df.dropna(subset='JobSat',axis=0)                    # dropping null values because over 50% of the column entries are empty so filling it with mean or median would disturb data
jobsat_df[['Employment','JobSat','RemoteWork']].isnull().sum()
```

```python
before_clean(jobsat_df['RemoteWork'])

most_frq_RW=jobsat_df['RemoteWork'].mode()[0]
jobsat_df['RemoteWork']=jobsat_df['RemoteWork'].fillna(most_frq_RW)       # Due to the small number of missing values, the most frequent category was used to impute RemoteWork.

print('most_frequent_RemoteWork_value:',most_frq_RW)
after_clean(jobsat_df['RemoteWork'])
```

#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
jobsat_df[['Employment','JobSat','RemoteWork','YearsCodePro']].describe(include='all')  # data check
```

```python
jobsat_df['YearsCodePro'].unique() # Identifying non-numeric values for data cleaning
```

```python
jobsat_df['YearsCodePro']=jobsat_df['YearsCodePro'].replace({'Less than 1 year': 0.5, 'More than 50 years': 55})
jobsat_df['YearsCodePro']=pd.to_numeric(jobsat_df['YearsCodePro'], errors='coerce')
years_median=jobsat_df['YearsCodePro'].median()
print('YearsCodePro median:', years_median)  # check
```

```python
before_clean(jobsat_df['YearsCodePro'])
jobsat_df['YearsCodePro']=jobsat_df['YearsCodePro'].fillna(years_median)        # Missing values in YearsCodePro were imputed using the median due to the small proportion of missing data
after_clean(jobsat_df['YearsCodePro'])
```

```python
jobsat_df[['Employment','JobSat','RemoteWork','YearsCodePro']].describe(include='all')
```

```python
## Write your code here
code_pro_range=[0,5,10,20,jobsat_df['YearsCodePro'].max()]
code_pro_names=['0-5','6-10','11-20','20+']
jobsat_df['code_pro_groups']=pd.cut(jobsat_df['YearsCodePro'], bins=code_pro_range, right=True, labels=code_pro_names, include_lowest=True)
ExpSat=jobsat_df.groupby('code_pro_groups')['JobSat'].median()
ExpSat.plot(kind='bar')
plt.xlabel('experience in years')
plt.ylabel('median job satisfaction in scale 1-10')
plt.title('work experience and job satisfaction')
plt.show()
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
cou_lang_df=df.copy()
print('shape', cou_lang_df.shape)  # check
```

```python
print('null values before dropping \n\n', cou_lang_df[['LanguageHaveWorkedWith','Country']].isnull().sum(),'\n\n\n')
cou_lang_df=cou_lang_df.dropna(subset=['LanguageHaveWorkedWith','Country'])                                                  # Rows with missing LanguageHaveWorkedWith and Country information were excluded from the analysis, as these variables are essential for the task.
print('null values after dropping \n\n', cou_lang_df[['LanguageHaveWorkedWith','Country']].isnull().sum())
```

```python
## Write your code here 

# Limit data to top 10 countries for better visualization clarity

top10_countries=(
    cou_lang_df.groupby('Country').size().reset_index(name='size')
    .sort_values('size', ascending=False).head(10) 
)

top10_countries=cou_lang_df[cou_lang_df['Country'].isin(top10_countries['Country'])]
top10_countries.shape  # check
```

```python
top10_countries['coding_lang']=top10_countries['LanguageHaveWorkedWith'].str.split(';')

results=(
    top10_countries.explode('coding_lang')
    .groupby(['Country','coding_lang']).size().reset_index(name='count')
)

results.shape  # check
```

```python
# Limit data to top 25 coding languages for better visualization clarity

top_cod_lang=(
    top10_countries.explode('coding_lang')
    .groupby('coding_lang').size().reset_index(name='count')
    .sort_values('count', ascending=False).head(25)
)
top_cod_lang.head()  #check
```

```python
results=results[results['coding_lang'].isin(top_cod_lang['coding_lang'])]
results.shape  # check
```

```python
pivot_coun_lang=results.pivot(index='coding_lang', columns='Country', values='count')
pivot_coun_lang.tail()
```

```python
pivot_coun_lang_norm=pivot_coun_lang/(pivot_coun_lang.sum(axis=0)) #Normalization was applied to avoid dominance of countries with large sample sizes in the heatmap.
```

```python
pivot_coun_lang_norm=pivot_coun_lang_norm.rename(columns={'United Kingdom of Great Britain and Northern Ireland':'UK'})
plt.figure(figsize=(20,24))
sns.heatmap(pivot_coun_lang_norm, cmap='Greens', annot=True, annot_kws={'fontsize':11, 'fontweight':'bold'})
plt.title('most popular programing languages in top 10 programing countries')
plt.show()
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
## Write your code here
edu_df=df.copy()
before_clean(edu_df['EdLevel'])
edu_df=edu_df.dropna(subset='EdLevel',axis=0)      #Rows with missing education level were removed, as this variable is important for the analysis and cannot be reliably imputed
after_clean(edu_df['EdLevel'])


edu_df['EdLevel'] = edu_df['EdLevel'].str.replace(r'\s*\(.*?\)', '', regex=True)


top_employment=(
    edu_df.groupby('Employment').size().reset_index(name='count')
    .sort_values('count', ascending=False).head(30)                       # Limit data to top 30 Employment for better visualization clarity
)

top_edu_df=edu_df[edu_df['Employment'].isin(top_employment['Employment'])]
top_edu_df.shape  # check
```

```python
education_employment=pd.crosstab(index=top_edu_df['Employment'], columns=top_edu_df['EdLevel'])
education_employment
```

```python
education_employment_norm=education_employment/(education_employment.sum(axis=0))    # Normalization was applied to prevent education levels with large sample sizes from dominating the heatmap.
```

```python
plt.figure(figsize=(20,24))
sns.heatmap(education_employment_norm, cmap='Greens', annot=True, annot_kws={'fontsize':11, 'fontweight':'bold'})
plt.title('most popular form of employment for each education level')
plt.show()
```
