### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
df_lang=df.copy()
print('NaN before droping \n\n',df_lang[['LanguageHaveWorkedWith', 'LanguageWantToWorkWith']].isnull().sum())

worked_lang = (
    df_lang['LanguageHaveWorkedWith'].dropna()
    .str.split(';')
    .explode()
    .value_counts()
)

wanted_lang = (
    df_lang['LanguageWantToWorkWith'].dropna()          #The two columns were analyzed independently to preserve all available information and avoid unnecessary loss of data due to missing values in only one of the fields.
    .str.split(';')
    .explode()
    .value_counts()
)

print('\n\n\n\n NaN after droping in worked_lang', worked_lang.isnull().sum(), '\t\t shape', worked_lang.shape)
print('\n\n NaN after droping in wanted_lang', wanted_lang.isnull().sum(), '\t\t shape', wanted_lang.shape)
```

```python
top15_lang=worked_lang[0:15].index                                       # Limit data to top 15 coding languages for better visualization clarity
wanted_lang_top15=wanted_lang[wanted_lang.index.isin(top15_lang)]
```

```python
wanted_lang_top15.head()
```

```python
top15_work_want=pd.concat([worked_lang[0:15], wanted_lang_top15], axis=1)   #Merging series
top15_work_want.columns=['worked','wanted']
top15_work_want
```

```python
top15_work_want.plot(kind='bar', figsize=(18,12))
plt.title('Comprassion of the frequency of LanguageHaveWorkedWith and LanguageWantToWorkWith')
plt.show()
```

note that there is 4,000 more answears for column LanguageHaveWorkedWith 

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

top20_countries=(
    df.groupby('Country').size().reset_index(name='size')
    .sort_values('size', ascending=False).head(20)                        # Limit data to top 20 countries for better visualization clarity
)

remote_country_df=df[df['Country'].isin(top20_countries['Country'])]
remote_country_df.shape # check
```

```python
print('NaN before imputation \n\n',remote_country_df[['Country', 'RemoteWork']].isnull().sum())
remote_country_df['RemoteWork']=remote_country_df['RemoteWork'].fillna('Unknown')                # Filling data with unknown will help find countries where respondents avoid answearing this question 
print('\n\n NaN after imputation ',remote_country_df['RemoteWork'].isnull().sum())
```

```python
remote_country_crosstab=pd.crosstab(index=remote_country_df['Country'], columns=remote_country_df['RemoteWork'])

remote_country_norm=remote_country_crosstab.div((remote_country_crosstab.sum(axis=1)), axis=0)            # Normalization was applied to avoid dominance of countries with large sample sizes in the heatmap.
remote_country_norm
```

```python
remote_country_norm=remote_country_norm.rename(index={'United Kingdom of Great Britain and Northern Ireland':'UK'})
plt.figure(figsize=(12, 20))
sns.heatmap(remote_country_norm, cmap='Greens', annot=True, annot_kws={'fontsize':11, 'fontweight':'bold'})
plt.title('distribution of RemoteWork by country')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
JobSat_and_exp=df.copy()
print('shape', JobSat_and_exp.shape)
print('\n NaN before cleaning \n\n',JobSat_and_exp[['JobSat', 'YearsCodePro']].isnull().sum())
JobSat_and_exp=JobSat_and_exp.dropna(subset=['JobSat', 'YearsCodePro'], axis=0)                        # Dropping rows with missing values in JobSat or YearsCodePro to ensure valid pairwise correlation.
print('\n\n\n\n NaN after cleaning \n\n',JobSat_and_exp[['JobSat', 'YearsCodePro']].isnull().sum())
print('\n shape', JobSat_and_exp.shape)
```

```python
JobSat_and_exp['YearsCodePro']=JobSat_and_exp['YearsCodePro'].replace({'Less than 1 year': 0.5, 'More than 50 years': 55})
JobSat_and_exp['YearsCodePro']=pd.to_numeric(JobSat_and_exp['YearsCodePro'], errors='coerce')
JobSat_and_exp[['JobSat', 'YearsCodePro']].info()
```

```python
plt.figure(figsize=(10,7))
sns.regplot(JobSat_and_exp, x='YearsCodePro', y='JobSat', line_kws={'color':'red'})
plt.title('correlation between overall job satisfaction and YearsCodePro')
plt.show()
```

```python
print('pearson correlation:', JobSat_and_exp[['JobSat', 'YearsCodePro']].corr())
print('\n spearman correlation:', JobSat_and_exp[['JobSat', 'YearsCodePro']].corr(method='spearman'))
```

Both Pearson and Spearman correlation coefficients indicate a very weak positive relationship between years of professional coding experience and overall job satisfaction

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
edu_df=df.copy()
print('NaN before cleaning',edu_df['EdLevel'].isnull().sum())
edu_df=edu_df.dropna(subset='EdLevel',axis=0)                                # Rows with missing education level were removed, as this variable is important for the analysis and cannot be reliably imputed
print('NaN after cleaning',edu_df['EdLevel'].isnull().sum())


edu_df['EdLevel'] = edu_df['EdLevel'].str.replace(r'\s*\(.*?\)', '', regex=True)


top_employment=(
    edu_df.groupby('Employment').size().reset_index(name='count')
    .sort_values('count', ascending=False).head(15)                       # Limit data to top 15 Employment for better visualization clarity
)

top_edu_df=edu_df[edu_df['Employment'].isin(top_employment['Employment'])]
top_edu_df.shape  # check
```

```python
education_employment=pd.crosstab(index=top_edu_df['Employment'], columns=top_edu_df['EdLevel'])
education_employment
```

```python
education_employment_norm = education_employment.div( # Normalaizing data to investigate education level in each employment group
    education_employment.sum(axis=1),axis=0)
```

```python
education_employment_norm.plot(kind='bar', stacked=True, figsize=(18,9))
plt.title('Relationship between employment status and education level')
plt.xticks(rotation=90)
plt.ylim(0,1.3)
plt.legend(loc='upper right', fontsize=8.5)
plt.show()
```
